```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
may24_16 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20160524_flow.shp')
jun10_16 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20160610_flow.shp')
jun30_16 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20160630_flow.shp')
jul19_16 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20160719_flow.shp')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))
[ax1, ax2, ax3, ax4] = axes.flatten()

may24_16.plot(ax=ax1)
ax1.set_title('May 24, 2016')

jun10_16.plot(ax=ax2)
ax2.set_title('June 10, 2016')

jun30_16.plot(ax=ax3)
ax3.set_title('June 30, 2016')

jul19_16.plot(ax=ax4)
ax4.set_title('July 19, 2016')

plt.show()
```

```python
nov29_16 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20161129_flow.shp')
dec14_16 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20161214_flow.shp')
jan12_17 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20170112_flow.shp')
feb24_17 = gpd.read_file('data/05_2016-17_kilauea_lava_shapefiles/Ep61g_20170224_flow.shp')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))
[ax1, ax2, ax3, ax4] = axes.flatten()

nov29_16.plot(ax=ax1)
ax1.set_title('November 29, 2016')

dec14_16.plot(ax=ax2)
ax2.set_title('December 14, 2016')

jan12_17.plot(ax=ax3)
ax3.set_title('January 12, 2017')

feb24_17.plot(ax=ax4)
ax4.set_title('February 24, 2017')

plt.show()
```
